```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV

# ----------- Mass Dataset -----------

X_mass = merged_mass.drop(columns=['pathology']) 
y_mass = merged_mass['pathology'] 

X_train_mass, X_test_mass, y_train_mass, y_test_mass = train_test_split(X_mass, y_mass, test_size=0.3, random_state=42)

# ----------- Random Forest for Mass Dataset -----------

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_mass, y_train_mass)

y_pred_rf_mass = rf.predict(X_test_mass)
rfma = accuracy_score(y_test_mass, y_pred_rf_mass)
print("Random Forest Accuracy for Mass Dataset:", rfma)
print("Random Forest Classification Report for Mass Dataset:\n", classification_report(y_test_mass, y_pred_rf_mass))

# ----------- XGBoost for Mass Dataset -----------

xgboost_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgboost_model.fit(X_train_mass, y_train_mass)

y_pred_xgb_mass = xgboost_model.predict(X_test_mass)
xgbma = accuracy_score(y_test_mass, y_pred_xgb_mass)
print("XGBoost Accuracy for Mass Dataset:", xgbma)
print("XGBoost Classification Report for Mass Dataset:\n", classification_report(y_test_mass, y_pred_xgb_mass))

# ----------- Calc Dataset -----------

X_calc = calc_merged.drop(columns=['pathology']) 
y_calc = calc_merged['pathology'] 

X_train_calc, X_test_calc, y_train_calc, y_test_calc = train_test_split(X_calc, y_calc, test_size=0.3, random_state=42)

# ----------- Random Forest for Calc Dataset -----------

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_calc, y_train_calc)

y_pred_rf_calc = rf.predict(X_test_calc)
rfca = accuracy_score(y_test_calc, y_pred_rf_calc)
print("Random Forest Accuracy for Calc Dataset:", rfca)
print("Random Forest Classification Report for Calc Dataset:\n", classification_report(y_test_calc, y_pred_rf_calc))

# ----------- XGBoost for Calc Dataset -----------

xgboost_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgboost_model.fit(X_train_calc, y_train_calc)

y_pred_xgb_calc = xgboost_model.predict(X_test_calc)
xgbca = accuracy_score(y_test_calc, y_pred_xgb_calc)
print("XGBoost Accuracy for Calc Dataset:", xgbca)
print("XGBoost Classification Report for Calc Dataset:\n", classification_report(y_test_calc, y_pred_xgb_calc))

# ----------- Overall Accuracy -----------

print('Overall Random Forest Accuracy:', (rfma + rfca) / 2)
print('Overall XGBoost Accuracy:', (xgbma + xgbca) / 2)

# ----------- Hyperparameter Tuning for Random Forest -----------

rf_param_dist = {
    'n_estimators': randint(100, 500),
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5),
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# ----------- Random Forest Hyperparameter Tuning for Mass Dataset -----------

print("Random Forest Hyperparameter Tuning for Mass Dataset:")
rf_random_search_mass = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=rf_param_dist,
    n_iter=100, cv=3, n_jobs=-1, verbose=2, random_state=42, scoring='accuracy'
)
rf_random_search_mass.fit(X_train_mass, y_train_mass)

print("Best Parameters for Mass Dataset Random Forest:", rf_random_search_mass.best_params_)
print("Best Accuracy for Mass Dataset Random Forest:", rf_random_search_mass.best_score_)

# ----------- Random Forest Hyperparameter Tuning for Calc Dataset -----------

print("Random Forest Hyperparameter Tuning for Calc Dataset:")
rf_random_search_calc = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=rf_param_dist,
    n_iter=100, cv=3, n_jobs=-1, verbose=2, random_state=42, scoring='accuracy'
)
rf_random_search_calc.fit(X_train_calc, y_train_calc)

print("Best Parameters for Calc Dataset Random Forest:", rf_random_search_calc.best_params_)
print("Best Accuracy for Calc Dataset Random Forest:", rf_random_search_calc.best_score_)

# ----------- Hyperparameter Tuning for XGBoost -----------

# XGBoost Parameter Distribution
xgb_param_dist = {
    'n_estimators': randint(100, 500),
    'max_depth': [3, 5, 7, 10],
    'learning_rate': uniform(0.01, 0.3),
    'subsample': uniform(0.7, 0.3),
    'colsample_bytree': uniform(0.7, 0.3),
    'gamma': uniform(0, 1),
    'reg_alpha': uniform(0, 0.1),
    'reg_lambda': uniform(0.1, 1.0)
}

# ----------- XGBoost Hyperparameter Tuning for Mass Dataset -----------

print("XGBoost Hyperparameter Tuning for Mass Dataset:")
xgb_random_search_mass = RandomizedSearchCV(
    estimator=xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    param_distributions=xgb_param_dist,
    n_iter=100, cv=3, n_jobs=-1, verbose=2, random_state=42, scoring='accuracy'
)
xgb_random_search_mass.fit(X_train_mass, y_train_mass)

print("Best Parameters for Mass Dataset XGBoost:", xgb_random_search_mass.best_params_)
print("Best Accuracy for Mass Dataset XGBoost:", xgb_random_search_mass.best_score_)

# ----------- XGBoost Hyperparameter Tuning for Calc Dataset -----------

print("XGBoost Hyperparameter Tuning for Calc Dataset:")
xgb_random_search_calc = RandomizedSearchCV(
    estimator=xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    param_distributions=xgb_param_dist,
    n_iter=100, cv=3, n_jobs=-1, verbose=2, random_state=42, scoring='accuracy'
)
xgb_random_search_calc.fit(X_train_calc, y_train_calc)

print("Best Parameters for Calc Dataset XGBoost:", xgb_random_search_calc.best_params_)
print("Best Accuracy for Calc Dataset XGBoost:", xgb_random_search_calc.best_score_)
```
